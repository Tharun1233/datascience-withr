```python
%%R
#create a vector
x <- rep(c("DS", "AI", "IOT", "MI"), c(8,10,5,2))
print(x)
#count number of occurences of each character string
mytable <- table(x)
print(mytable)
#bar plot using that table
barplot(mytable)
```

```python
%%R
barplot(mytable,col=1:4,maim="barplot",las=0)
```

```python
%%R
#create a vector
x<- rep(c("IOT", "ML", "DS", "AI"),c(8,10,5,2))

#count numbersof occurences of each string 
mytable <- table(x)

pie(mytable,
    main="The pie chart",
    col=c("lightblue", "lightgreen", "salmon", "maroon"))
```

```python
%%R
#create a matrix of 1000 random values from the normal distribution
x <-  matrix(rnorm(1000), ncol=4)
xdf=as.data.frame(x)
print(xdf)
# basic boxplot
boxplot(xdf)
```

```python
%%R
# vector of 200 random values from normal distribution
x <- rnorm(200)

# plot histogram
hist(x)
print(mean(x))
print(sd(x))
```

```python
%%R
hist(x,
     border="blue",
     breaks=50,
     main="histogram",
     xlab="",
     cex.main=2.5,
     cex.lab=2)
```

```python
%%R
library(ggplot2)
```

```python
%%R
data("iris")
ggplot(data=iris, mapping=aes(x=Species, y=Sepal.Length))
```

```python
%%R
data("iris")
ggplot(data=iris, mapping=aes(x=Species, y=Sepal.Length))+ geom_point()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length))+ geom_point()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length))+ geom_point(color="red")
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length,color=Species))+
geom_point()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length,fill=Species))+
geom_boxplot()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_bar()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_histogram()
```

```python
%%R 
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_histogram(position="dodge")
```
